```python
# 確保目標欄位存在
target_column = '📌 Part 1: 了解 API\n什麼是 API？請用自己的話解釋'
if target_column not in df.columns:
    print(f"錯誤：欄位 '{target_column}' 不存在於試算表中。")
else:
    # 為 Jieba 添加詞語 (如果需要)
    for _, row in df.iterrows():
        word = row[target_column]
        if isinstance(word, str):  # 確保 word 是字串
            jieba.add_word(word, freq=10000)
            jieba.suggest_freq(word, tune=True)

    # 只保留中文字
    def extract_chinese(text):
        if isinstance(text, str):
            return ''.join(re.findall(r'[\u4e00-\u9fa5]+', text))
        return ''

    df['Chinese_Content'] = df[target_column].apply(extract_chinese)

    # 使用 Jieba 斷詞
    df['Tokenized'] = df['Chinese_Content'].apply(lambda x: list(jieba.cut(x, HMM=True)))

    # 實體提取
    def extract_entities(tokens):
        entities = []

        for token in tokens:
            if token == "API":
                entities.append({"text": token, "type": "Concept"})
        return entities

    df['Entities'] = df['Tokenized'].apply(extract_entities)

    print(df[['Chinese_Content', 'Tokenized', 'Entities']].to_string())

    # 結果儲存為 JSON 格式的列表
    output_list = []
    for _, row in df.iterrows():
        output_list.append({
            "text": row['Chinese_Content'],
            "tokens": row['Tokenized'],
            "entities": row['Entities']
        })

    import json
    print("\n輸出 JSON 格式:")
    print(json.dumps(output_list[:5], ensure_ascii=False, indent=2))
```
