```python
# 開啟 Google Sheet
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/103Q1qgG_RKBX3Bf4jWmWO8FZP6wePTu33eRLMUwOUlI/edit?usp=sharing')
worksheet = gsheets.get_worksheet(0)

# 取得所有資料
rows = worksheet.get_all_values()

# 自訂欄位名稱（略過空欄）
headers = ['編號', '學號', '學校班級', '姓名', '錄取學校', '國文', '英文', '數A', '數B', '社會', '自然', '英聽', '入取管道']

# 因為前面有 3 個空欄，所以要取 index 的方式來對應正確欄位
data_rows = []
for row in rows[1:]:  # 從第 2 列開始是資料
    # 組合正確欄位（跳過 index 5~7 的空值）
    filtered_row = row[0:5] + row[8:]
    data_rows.append(filtered_row)

# 轉為 DataFrame
df = pd.DataFrame(data_rows, columns=headers)
```

```python
# 欄位標題（略過空白欄）
headers = ['編號', '學號', '學校班級', '姓名', '錄取學校', '國文', '英文', '數A', '數B', '社會', '自然', '英聽', '入取管道']

# 過濾空欄（index 5~7）後取出資料
data_rows = []
for row in rows[1:]:
    filtered_row = row[0:5] + row[8:]
    data_rows.append(filtered_row)

# 建立 DataFrame
df = pd.DataFrame(data_rows, columns=headers)

# 要清洗的成績欄位
score_columns = ['國文', '英文', '數A', '數B', '社會', '自然']

# 移除括號與內文，只保留數字
for col in score_columns:
    df[col] = df[col].str.extract(r'(\d+)')  # 抓出數字部分
    df[col] = pd.to_numeric(df[col], errors='coerce')  # 轉成數值型態，非數字轉 NaN

# 顯示前幾筆清洗後的資料
df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import plotly.express as px
```

```python
# 做 KMeans 聚類（假設你想分成 3 群）
kmeans_model = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_features = df[score_columns].dropna()
df_valid = df.loc[cluster_features.index]
df_valid['cluster'] = kmeans_model.fit_predict(cluster_features)

# PCA 降維
pca = PCA(n_components=2)
pca_result = pca.fit_transform(cluster_features)
df_valid['pca1'] = pca_result[:, 0]
df_valid['pca2'] = pca_result[:, 1]

# 畫圖
fig = px.scatter(
    df_valid, x='pca1', y='pca2',
    color=df_valid['cluster'].astype(str),
    hover_data=['姓名', '學校班級', '錄取學校'],
    title='學生成績 KMeans 聚類 + PCA 可視化'
)
fig.show()
```

# 學生成績 KMeans 聚類 + PCA 可視化（含各科成績）

🔵 藍色	0	成績分布較平均或居中，主成分值落在右側，有可能是「表現優異」的群體。例如:海山309  

🟢 綠色	1	多數集中在左下方，PCA 值偏負，可能代表「成績整體較低」的群體。如三重三律

🔴 紅色	2	分布在中間偏右，有可能代表「某幾科特別強或偏科」的學生群。如新莊316








各群趨勢細節

✅ 紅色群組（color=2）：

明顯集中在圖的右中區。

代表這些學生的 PCA 主成分分數較高，某些成績特別突出。

### 看起來這群學生的成績變異度較大，可能是某些科目特別強（如自然或數學A高分）。





✅ 綠色群組（color=1）：

集中在圖的左側偏下。pca1 和 pca2 多為負值，表示整體成績較低或在主成分上無優勢。

### 這群學生可能是整體學科表現偏弱的群體。





✅ 藍色群組（color=0）：

分散程度較大，但多數集中在右下方。主成分值整體偏正，成績可能 均衡且中上程度。

### 這群學生可能是「穩定發揮」的類型，各科不偏科。

```python
fig = px.scatter(
    df_valid,
    x='pca1', y='pca2',
    color=df_valid['cluster'].astype(str),
    hover_data=[
        '學校班級', '錄取學校',
        '國文', '英文', '數A', '數B', '社會', '自然'  # 加入科目成績
    ],
    title='學生成績 KMeans 聚類 + PCA 可視化（含各科成績）'
)
fig.show()
```

 圖表解讀：以「國文」與「英文」為基礎的 KMeans 分群散點圖

 軸向與變數：

X 軸：國文成績

Y 軸：英文成績

顏色：KMeans 分群（0：藍、1：綠、2：紅）


群 0（藍）	國文成績高，英文成績也高

群 1（綠）	國文成績中等，英文成績較低

群 2（紅）	國文成績中～高，英文成績中等

🧠 各群組的趨勢分析：

🔵 群 0（藍色）：
分布在 國文 12到14 級，英文 11到14 級 的區域。

特徵：國文與英文都表現非常好。

**這群學生屬於「雙語強者」，語文能力明顯突出，是整體表現最佳的一群。**

🟢 群 1（綠色）：
分布集中在 國文 8到11 級，英文 4到7 級。

特徵：國文中等，英文偏低。

**這群學生可能偏好中文學習，英文為相對弱項，屬於「中語文、弱英文型」。**

🔴 群 2（紅色）：
分布在 國文 9到13 級，英文 5到10 級。

特徵：國文略高於英文，但兩者差距不如群 1 明顯。

**屬於「語文平均型」，語文整體普通，未特別強弱。**
